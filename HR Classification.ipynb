```python
import pandas as pd 
import numpy as np 
import plotly.express as px
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv("HRDataset_v14.csv")
data.columns
```

```python
data_new = data[["Employee_Name", "EmpID", "GenderID", "Salary", "Position", "Zip", "MaritalDesc", "DateofHire", 
                 "DateofTermination", "TermReason", "Department", "EmploymentStatus", "ManagerName", 
                 "SpecialProjectsCount", "LastPerformanceReview_Date", "DaysLateLast30"]].set_index("EmpID")
data_new.Position.unique()
```

```python
ohe = OneHotEncoder()
ohe.fit(data_new[["ManagerName"]])
manager_enc = pd.DataFrame(ohe.transform(data_new[["ManagerName"]]).toarray())
manager_enc
```
